```python
import imutils # 파이썬 OpenCV가 제공하는 기능 중 복잡하고 사용성이 떨어지는 부분을 보완(이미지 또는 비디오 스트림 파일 처리 등)
import cv2 # opencv 모듈
```

# gray_resize_vid

```python
def gray_resize_vid(video_path):
  print("[video 시작]")
  vs = cv2.VideoCapture(video_path)

  # 비디오 저장 변수
  writer = None

  # 비디오 스트림 프레임 반복
  while True:
    # 프레임 읽기
    ret, frame = vs.read()

    # 읽은 프레임이 없는 경우 종료
    if frame is None:
        break
    
    # 프레임 resize
    frame = imutils.resize(frame, width=500)
    frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)

    
    key = cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        break
                                    
    # 저장할 비디오 설정
    if writer is None:
        fourcc = cv2.VideoWriter_fourcc(*"MJPG")
        writer = cv2.VideoWriter(result_path, fourcc, 25, (frame.shape[1], frame.shape[0]),False)
    #####RGB==True, non==False설정 주의


    # 비디오 저장
    if writer is not None:
        writer.write(frame)

  # 종료
  vs.release()
  cv2.destroyAllWindows()
```

```python
video_path= "./gyro_swing_1.mp4"
result_path = "./result_video.avi"

gray_resize_vid(video_path)
```

# 검수용 caption 비디오


```python
import cv2

def caption_vid(video_path):
  print("[video 시작]")
  vs = cv2.VideoCapture(video_path)
  writer = None
  cnt=0

  while True:
    ret, frame = vs.read()

    if frame is None:
        break
    frame=cv2.putText(frame, 'frame: '+str(cnt), (0, 100), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0))
    cnt+=1
    
    key = cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        break
                                    
    # 저장할 비디오 설정
    if writer is None:
        fourcc = cv2.VideoWriter_fourcc(*"MJPG")
        writer = cv2.VideoWriter('captioned_gyro_swing1.avi', fourcc, 25, (frame.shape[1], frame.shape[0]),True)
    #####RGB==True, non==False설정 주의


    # 비디오 저장
    if writer is not None:
        writer.write(frame)

  # 종료
  vs.release()
  cv2.destroyAllWindows()



caption_vid('/content/drive/MyDrive/amusementCycle/gyro_swing1.mp4')
```
